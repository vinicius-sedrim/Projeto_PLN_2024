### **MODELO**
---




**Objetivo do Modelo**


Configurar o Nous-Hermes-13B-GPTQ, um modelo de linguagem de grande escala (LLM), otimizado para execução em GPUs com suporte à quantização, para processar e responder a perguntas em tempo real usando técnicas avançadas de Processamento de Linguagem Natural (PLN). Este modelo é essencial para gerar respostas inteligentes e contextualizadas, simulando a interação como um atendente humano.

**Por Que Precisamos deste Modelo?**
- **Compreensão de Linguagem Natural**: Capacidade de entender intenções complexas nas perguntas dos usuários e fornecer respostas coerentes.
- **Geração de Respostas**: Produzir respostas detalhadas que simulam conversas humanas.
- **Escalabilidade**: Processamento eficiente de grandes volumes de dados através do uso de GPUs e técnicas de quantização.

**Necessidade de GPUs e Quantização**
- **GPUs**: Aceleram o processamento de modelos de linguagem complexos.
- **Quantização**: Reduz a precisão dos cálculos para otimizar a performance sem comprometer significativamente a qualidade.

**Processo de Inferência**
1. **Tokenização**: Conversão do texto da pergunta em IDs numéricos que o modelo pode processar.
2. **Geração de Resposta**: Utilização de técnicas de controle de temperatura e penalidade de repetição para produzir respostas variadas e coerentes.
3. **Decodificação**: Transformação das saídas numéricas do modelo de volta para texto legível.

**Integração com API Hugging Face**

Uso da API para carregar o Nous-Hermes-13B-GPTQ e o tokenizador correspondente, acessando uma vasta gama de modelos pré-treinados para geração de texto e respostas.

**Considerações**

A configuração e utilização do Nous-Hermes-13B-GPTQ, juntamente com as ferramentas fornecidas pela Hugging Face, oferecem uma solução robusta e escalável para o chatbot, permitindo gerenciar eficientemente interações complexas e volumes de dados significativos.


```python
DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
# Define o nome e o caminho do modelo a ser carregado
model_name_or_path = "TheBloke/Nous-Hermes-13B-GPTQ"
use_triton = False  # Usar Triton se necessário para otimizar a execução

# Carrega o tokenizador associado ao modelo para transformar texto em IDs numéricos
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path, use_fast=True)

# Carrega o modelo otimizado para quantização, possibilitando execução em GPUs
model = AutoGPTQForCausalLM.from_quantized(
    model_name_or_path,
    use_safetensors=True,  # Usa formato seguro para carregar pesos do modelo
    trust_remote_code=True,  # Confia no código remoto para configuração customizada
    device= DEVICE,  # Aloca o modelo para a GPU
    #device= "cpu", # Carrega o modelo na CPU
    use_triton=use_triton,  # Usa Triton se necessário para otimização
    quantize_config=None  # Nenhuma configuração de quantização adicional
)

# Configuração de geração baseada no modelo pré-carregado
generation_config = GenerationConfig.from_pretrained(model_name_or_path)
```

```python
# Exemplo de pergunta para ser respondida pelo modelo
question = (
    "Which programming language is more suitable for a beginner: Python or JavaScript?"
)
prompt = f"""
### Instruction: {question}
### Response:
""".strip()
```

```python
%%time
# Geração de IDs de entrada a partir do texto do prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(DEVICE)

# Executa o modelo em modo de inferência, sem cálculo de gradientes
with torch.inference_mode():
    output = model.generate(inputs=input_ids, temperature=0.7, max_new_tokens=512)
```

```python
# Decodifica a saída numérica de volta para texto legível
print(tokenizer.decode(output[0]))
```

```python
generation_config
```

```python
# Configura um streamer para exibir a saída do modelo
streamer = TextStreamer(
    tokenizer, skip_prompt=True, skip_special_tokens=True, use_multiprocessing=False
)
```

```python
# Define um pipeline de geração de texto para o modelo, incluindo configurações adicionais
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=2048, # Número máximo de tokens gerados
    temperature=0, # Controla a aleatoriedade na geração de texto
    top_p=0.95, # Controla a diversidade do texto gerado
    repetition_penalty=1.15, # Evita repetições no texto gerado
    generation_config=generation_config, # Utiliza as configurações pré-definidas de geração
    streamer=streamer, # Exibe as respostas à medida que são geradas
    batch_size=1, # Gera uma resposta por vez
)
```

```python
# Integra o modelo no pipeline do HuggingFace
from langchain_huggingface import HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
# Gera uma resposta baseada no prompt
response = llm(prompt)
```

### **INCORPORAR DOCUMENTOS**
---


**Objetivo**

Habilitar o chatbot a responder perguntas baseadas em documentos armazenados, usando embeddings para realizar buscas semânticas eficientes. Isso permite que o sistema encontre respostas relevantes mesmo quando a consulta do usuário não é uma correspondência exata de palavras, mas sim de conceitos.

**Importância dos Embeddings**

- **Busca Semântica**: Permite ao chatbot procurar conteúdo conceitualmente similar à pergunta do usuário, melhorando a precisão das respostas.
- **Eficiência**: Acelera a recuperação de documentos relevantes usando representações vetoriais.
- **Escalabilidade**: Facilita o gerenciamento de grandes volumes de dados, tornando o sistema mais ágil.

**Componentes da Incorporação**
- **Modelo de Embedding**: Utilização do `sentence-transformers-multilingual-e5-base` da Hugging Face para converter documentos em vetores numéricos, facilitando cálculos de similaridade.
- **Carregamento de Documentos**: Os textos são carregados e convertidos em objetos Document, que incluem conteúdo e metadados para facilitar referências futuras.
- **Divisão de Textos**: Documentos longos são segmentados em blocos de 512 caracteres para melhorar a eficiência do processamento.
- **Armazenamento Vetorial**: O banco de dados vetorial Chroma armazena os embeddings, permitindo buscas rápidas por similaridade.
- **Busca por Similaridade**: Implementação da função `similarity_search()` para encontrar documentos que correspondam semanticamente às consultas dos usuários.

**Necessidade de Embeddings**

Incorporar documentos como embeddings possibilita uma compreensão mais profunda das perguntas dos usuários e uma busca de informações mais natural e inteligente. Esta abordagem é vital para lidar com um grande número de documentos e responder a consultas de forma eficaz e precisa.

**Uso da API Hugging Face**

A API da Hugging Face é utilizada para gerar embeddings dos documentos carregados, facilitando a realização de buscas semânticas em vários idiomas através do modelo `sentence-transformers-multilingual-e5-base`.

**Considerações**

O uso de embeddings otimiza significativamente o sistema de busca do chatbot, permitindo que ele forneça respostas detalhadas e precisas com base em um extenso conjunto de dados, melhorando a experiência do usuário.


```python
# Carrega o modelo de embeddings HuggingFace para transformar textos em vetores numéricos
embeddings = HuggingFaceEmbeddings(
    model_name="embaas/sentence-transformers-multilingual-e5-base",
    model_kwargs={"device": DEVICE}, # Define o dispositivo (GPU ou CPU)
)
```

```python
# Carrega os documentos do diretório 'skyscanner'
loader = DirectoryLoader("./skyscanner/", glob="**/*txt")
documents = loader.load()
len(documents)
```

```python
# Divide os documentos em partes menores para facilitar o processamento
text_splitter = CharacterTextSplitter(chunk_size=512, chunk_overlap=0)
texts = text_splitter.split_documents(documents)
```

```python
texts[4] # Visualiza um trecho do documento dividido
```

```python
# Cria um banco de dados vetorial com base nos textos divididos e seus embeddings
db = Chroma.from_documents(texts, embeddings)
```

```python
# Realiza uma busca por similaridade semântica nos documentos carregados
db.similarity_search("flight search")
```

### **CADEIA DE CONVERÇÃO**

**Objetivo da Cadeia de Conversação**

Implementar um mecanismo que permite ao chatbot manter o contexto ao longo de múltiplas interações com o usuário, simulando uma troca fluida e contínua como entre um agente de suporte e um cliente. Isso envolve o uso do histórico de conversa para gerar respostas relevantes e informativas.

**Importância da Cadeia de Conversação**
- **Manutenção do Contexto**: O sistema acompanha a sequência de perguntas e respostas para não perder o contexto da conversa.
- **Respostas Mais Precisas**: Utiliza o histórico para adaptar as respostas, tornando o suporte mais informado e relevante.
- **Simulação de Conversas Reais**: A cadeia de conversação torna o chatbot mais humano e empático, melhorando significativamente a experiência do usuário.

**Necessidade da Cadeia de Conversação**
- **Experiência de Suporte Natural**: Evita que usuários precisem repetir informações, proporcionando uma experiência mais natural e fluida.

### **CHATBOT DE SUPORTE**
---

**Objetivo do Chatbot de Suporte**

O chatbot visa simular interações humanas de forma eficiente, atuando como agente de suporte ao cliente. Ele é projetado para responder perguntas, analisar sentimentos, fornecer resumos de texto, e utilizar dados e histórico de conversas para oferecer respostas personalizadas e contextuais.

**Necessidade do Chatbot**
- **Automatização do Suporte ao Cliente**: Reduz a necessidade de interação humana ao responder autonomamente a uma vasta gama de perguntas.
- **Geração de Respostas Contextuais**: Utiliza memória e documentos incorporados para fornecer respostas precisas e relevantes.
- **Personalização da Experiência**: Adapta respostas com base em interações anteriores, criando uma experiência personalizada para cada usuário.

**Componentes Principais**
- **Carregamento de Documentos**: Utiliza a função `_embed_data` para acessar textos que servem de base para as respostas.
- **Template de Resposta**: Define como as respostas são formuladas, assegurando que sejam empáticas e informativas.
- **Cadeia de Perguntas e Respostas**: Integra memória de conversação e busca documental para responder de maneira contextualizada.
- **Funcionalidades Adicionais**: Inclui análise de sentimentos e sumarização de textos, ativadas conforme a necessidade do usuário.
- **Interação com o Usuário**: O chatbot interage diretamente com o usuário, executando tarefas específicas e respondendo perguntas até o término da conversa.

**Utilização da API Hugging Face**

A API da Hugging Face é essencial, utilizando o `HuggingFacePipeline` para conectar o modelo de linguagem ao chatbot, permitindo a geração de texto e a execução de inferências. A API também é usada para gerar embeddings que facilitam a busca por informações nos documentos incorporados.

**Considerações**

O Chatbot de Suporte combina várias tecnologias de Processamento de Linguagem Natural para fornecer uma interação interativa e valiosa, garantindo suporte ao cliente ágil, preciso e personalizado. Além de responder perguntas, o chatbot enriquece a experiência do usuário com funcionalidades avançadas como análise de sentimentos e sumarização de textos.


```python
DEFAULT_TEMPLATE = """
### Instruction: You're a travelling support agent that is talking to a customer. Use only the chat history and the following information
{context}
to answer in a helpful manner to the question. If you don't know the answer - say that you don't know.
Keep your replies short, compassionate and informative.
{chat_history}
### Input: {question}
### Response:
""".strip()


class Chatbot:
    def __init__(
        self,
        text_pipeline: HuggingFacePipeline,  # Pipeline de processamento de texto para gerar respostas.
        embeddings: HuggingFaceEmbeddings,  # Embeddings para representar os documentos como vetores.
        documents_dir: Path,  # Diretório onde os documentos são armazenados.
        prompt_template: str = DEFAULT_TEMPLATE,  # Template de prompt usado para gerar respostas.
        verbose: bool = False,  # Flag para ativar a saída detalhada (opcional).
    ):
        # Criação do template de prompt com variáveis de entrada.
        prompt = PromptTemplate(
            input_variables=["context", "question", "chat_history"],
            template=prompt_template,
        )
        # Cria a cadeia de processamento com o pipeline de texto e o template de prompt.
        self.chain = self._create_chain(text_pipeline, prompt, verbose)
        # Cria o banco de dados de documentos com embeddings.
        self.db = self._embed_data(documents_dir, embeddings)

    def _create_chain(
        self,
        text_pipeline: HuggingFacePipeline,  # Pipeline de processamento de texto para gerar respostas.
        prompt: PromptTemplate,  # Template de prompt para formatar perguntas e respostas.
        verbose: bool = False,  # Flag para ativar a saída detalhada (opcional).
    ):
        # Configuração da memória para armazenar o histórico de conversas.
        memory = ConversationBufferMemory(
            memory_key="chat_history",  # Chave para o histórico de conversa.
            human_prefix="### Input",  # Prefixo para as entradas do usuário.
            ai_prefix="### Response",  # Prefixo para as respostas do chatbot.
            input_key="question",  # Chave para a pergunta.
            output_key="output_text",  # Chave para a resposta gerada.
            return_messages=False,  # Não retorna mensagens completas, apenas o texto.
        )

        # Cria a cadeia de QA (questão e resposta) usando o pipeline de texto e a memória configurada.
        return load_qa_chain(
            text_pipeline,
            chain_type="stuff",  # Tipo de cadeia de processamento (pode ser ajustado conforme necessário).
            prompt=prompt,
            memory=memory,
            verbose=verbose,
        )

    def _embed_data(
        self, documents_dir: Path, embeddings: HuggingFaceEmbeddings  # Diretório de documentos e embeddings para representação vetorial.
    ) -> Chroma:
        # Carrega os documentos do diretório especificado.
        loader = DirectoryLoader(documents_dir, glob="**/*txt")
        documents = loader.load()
        # Divide os documentos em pedaços menores para processamento.
        text_splitter = CharacterTextSplitter(chunk_size=512, chunk_overlap=0)
        texts = text_splitter.split_documents(documents)
        # Cria um banco de dados de vetores (embeddings) a partir dos textos divididos.
        return Chroma.from_documents(texts, embeddings)

    def __call__(self, user_input: str) -> str:
        # Realiza uma busca de similaridade no banco de dados usando a entrada do usuário.
        docs = self.db.similarity_search(user_input)
        # Gera uma resposta baseada nos documentos encontrados e na pergunta do usuário.
        return self.chain.run({"input_documents": docs, "question": user_input})
```

```python
# Cria uma instância do chatbot com o pipeline de linguagem (llm), embeddings para documentos e o diretório de documentos.
chatbot = Chatbot(llm, embeddings, "./skyscanner/")
```

```python
import warnings

# Ignora os avisos de categoria UserWarning para evitar que mensagens não críticas apareçam durante a execução.
warnings.filterwarnings("ignore", category=UserWarning)

# Loop principal para interação com o usuário.
while True:
    # Solicita a entrada do usuário.
    user_input = input("You: ")

    # Verifica se a entrada do usuário é uma mensagem de despedida para encerrar o loop.
    if user_input.lower() in ["bye", "goodbye"]:
        break

    # Passa a entrada do usuário para o chatbot e obtém a resposta.
    answer = chatbot(user_input)

    # Imprime a resposta do chatbot.
    print()
```

### **SUMARIZAÇÃO**
---



**Funcionalidades Principais**
- **Criação e Configuração do Chatbot**: Inicia com um pipeline para processamento de texto e embeddings, acessando documentos armazenados especificamente.
- **Processamento de Texto e Respostas**: Gera respostas utilizando um template de prompt, baseando-se nas perguntas dos usuários e no histórico de conversas.
- **Memória e Cadeia de Processamento**: Armazena histórico de conversas para manter o contexto e usa uma cadeia de perguntas e respostas para responder com relevância e precisão.
- **Indexação e Pesquisa de Documentos**: Processa documentos para criar um banco de dados vetorial, permitindo buscas por similaridade para encontrar conteúdo relevante.

**Função Específica de Sumarização**
- **Objetivo**: Condensar textos longos em resumos mais curtos e compreensíveis.
- **Funcionamento**: Utiliza um prompt específico para sumarizar textos recebidos, facilitando a compreensão do usuário.
- **Aplicação**: Ideal para resumir informações detalhadas, como descrições de pacotes de viagem, destacando os pontos principais.

**Benefícios da Função de Sumarização**

A sumarização não apenas economiza tempo ao apresentar informações condensadas, mas também torna o conteúdo mais acessível, melhorando a experiência do usuário ao interagir com o chatbot.


```python
DEFAULT_TEMPLATE = """
### Instruction: You're a travelling support agent that is talking to a customer. Use only the chat history and the following information
{context}
to answer in a helpful manner to the question. If you don't know the answer - say that you don't know.
Keep your replies short, compassionate and informative.
{chat_history}
### Input: {question}
### Response:
""".strip()

class Chatbot:
    def __init__(
        self,
        text_pipeline: HuggingFacePipeline,  # Pipeline para processar o texto com o modelo de linguagem
        embeddings: HuggingFaceEmbeddings,   # Embeddings para representar documentos de forma vetorial
        documents_dir: Path,                 # Diretório contendo os documentos a serem indexados
        prompt_template: str = DEFAULT_TEMPLATE,  # Template padrão para a formatação de prompts
        verbose: bool = False,               # Flag para ativar/desativar mensagens detalhadas de debug
    ):
        # Cria o template de prompt usando variáveis de entrada
        prompt = PromptTemplate(
            input_variables=["context", "question", "chat_history"],
            template=prompt_template,
        )
        # Cria a cadeia de processamento usando o modelo e o template de prompt
        self.chain = self._create_chain(text_pipeline, prompt, verbose)
        # Indexa os documentos e cria o banco de dados de vetores
        self.db = self._embed_data(documents_dir, embeddings)

    def _create_chain(
        self,
        text_pipeline: HuggingFacePipeline,  # Pipeline para processar o texto com o modelo de linguagem
        prompt: PromptTemplate,              # Template de prompt usado para gerar respostas
        verbose: bool = False,              # Flag para mensagens detalhadas de debug
    ):
        # Configura a memória para armazenar o histórico de conversa
        memory = ConversationBufferMemory(
            memory_key="chat_history",         # Chave para acessar o histórico de conversa
            human_prefix="### Input",          # Prefixo para mensagens do usuário
            ai_prefix="### Response",          # Prefixo para respostas do chatbot
            input_key="question",              # Chave para a pergunta do usuário
            output_key="output_text",         # Chave para a resposta gerada
            return_messages=False,             # Não retorna as mensagens armazenadas
        )

        # Cria e retorna a cadeia de QA configurada
        return load_qa_chain(
            text_pipeline,
            chain_type="stuff",                # Tipo de cadeia de processamento
            prompt=prompt,
            memory=memory,
            verbose=verbose,
        )

    def _embed_data(
        self, documents_dir: Path,             # Diretório contendo os documentos a serem indexados
        embeddings: HuggingFaceEmbeddings     # Embeddings para representar os documentos
    ) -> Chroma:
        # Carrega os documentos do diretório especificado
        loader = DirectoryLoader(documents_dir, glob="**/*txt")
        documents = loader.load()
        # Divide os documentos em pedaços menores para processamento
        text_splitter = CharacterTextSplitter(chunk_size=512, chunk_overlap=0)
        texts = text_splitter.split_documents(documents)
        # Cria um banco de dados de vetores a partir dos textos
        return Chroma.from_documents(texts, embeddings)

    def __call__(self, user_input: str) -> str:
        # Realiza uma busca por similaridade nos documentos indexados
        docs = self.db.similarity_search(user_input)
        # Gera uma resposta para a pergunta do usuário com base nos documentos encontrados
        return self.chain.run({"input_documents": docs, "question": user_input})

    # Função para sumarização de texto
    def summarize_text(self, text):
        # Cria um prompt para solicitar a sumarização do texto fornecido
        prompt_text = f"Summarize the following text: '{text}'."
        # Gera e retorna o resumo do texto
        summary = self.chain.run({"input_documents": [], "question": prompt_text})
        return summary.strip()

# Cria uma instância do chatbot com o modelo de linguagem, embeddings e diretório de documentos
chatbot = Chatbot(llm, embeddings, "./skyscanner/")

import warnings

# Ignora avisos de categoria UserWarning para evitar mensagens não críticas
warnings.filterwarnings("ignore", category=UserWarning)

# Loop principal para interação com o usuário
while True:
    # Solicita a entrada do usuário
    user_input = input("You: ")
    # Verifica se a entrada do usuário é uma mensagem de despedida para encerrar o loop
    if user_input.lower() in ["bye", "goodbye"]:
        break
    # Passa a entrada do usuário para o chatbot e obtém a resposta
    answer = chatbot(user_input)
    # Imprime a resposta do chatbot
    print()
```
